## Arreglos

```python
# Arreglos (Arrays)
import numpy as np

np.array([1,2,3])

# Manjejo matemático

np.array([1,2,3]) + np.array([2,3,4])
```

```python
# Tamaño Fijo
np.ones([5,3])
```

```python
Array = np.zeros([4,7])
```
